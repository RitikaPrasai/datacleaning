```python
def season1(month):
    if month==12 or month==1 or month==2:
        return "winter"   
    elif month==3 or month==4 or month==5 :
        return "spring"
    elif month==6 or month==7 or month==8:
        return "summer"
    else:
        return "fall"
```

```python
season_copy=new_climatedata_copy.copy()
```

```python
def getMonths(input, m1, m2, m3):
    return input.loc[(input.date.dt.month==m1) | (input.date.dt.month==m2) | (input.date.dt.month==m3)]
```

```python
Spring = getMonths(season_copy,3,4,5)
```

```python
season_list = []

for month in season_copy['MO']:
    season = season1(month)
    season_list.append(season)
    
season_copy['Season_list'] = season_list
```
